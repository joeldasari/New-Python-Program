# Task 1
* <b> (2,9,8,5,6,7,8,9,10,2)
* create a tuple with 10 numbers
* print list of squares of elements in the tuple
* convert list to tuple
* find the max element index position of the tuple 
* find the count min value

```python
t=(2,9,8,5,6,7,8,9,10,2)
m=[]
for i in t:
    m.append(i*i)
#print(m)
m=tuple(m)
print(m)
print(m.index(max(m)))
print(m.count(min(m)))
```
